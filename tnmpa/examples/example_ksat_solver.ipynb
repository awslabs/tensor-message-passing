```python
from vectorized_tensorKSAT import KsatInstance, KsatSolver
```

```python
import random
import numpy as np
```

```python
# paramter the controls the complexity of the problem, ratio between number of clauses and number of variables
# for alpha > 3.5 BP should start to fail
alpha = 3.5 

# number of variables
N = 500

# numer of clauses
M = int(alpha * N)

# K of K-SAT
K = 3
```

```python
seed = 1
random.seed(seed)
np.random.seed(seed)

instance = KsatInstance(N, M, K)
```

```python
ksol = KsatSolver(instance, solver="BP")
```

```python
tol = 1e-2
ksol.solve(tol=tol, backend="numpy")
```
